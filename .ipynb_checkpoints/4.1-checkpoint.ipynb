```python
import numpy as np
import matplotlib.pyplot as plt

def fx(x):
    return 1/(1+x**2) 

def lagrange(f,start,end,step,x_test):
    interval_num = int((end - start)/step)+1
    x = np.linspace(start,end,interval_num,endpoint=True)

    L = 0
    for x_i in x:
        l_x = 1
        for x_j in x:
            if(x_i!=x_j):
                l_x = l_x*((x_test - x_j)/(x_i-x_j))
        L = L + l_x*f(x_i)
    return L,x
   
plt.figure(figsize=(6, 6),dpi=200)
plt.rcParams['axes.unicode_minus']=False

for index,step in enumerate([-1,2,1,0.5]):
    ax = plt.subplot(2,2,index+1)
    x = np.arange(-5,5.1,0.1)
    if(step == -1):
        y = fx(x)
    else:
        y,x_p = lagrange(fx,-5,5,step,x)
        y_p,x_p = lagrange(fx,-5,5,step,x_p)
        plt.scatter(x_p,y_p,color='blue') 
        y_true = fx(x)
        plt.plot(x,y_true, color="grey",linestyle = "--")
    plt.plot(x,y, color="r",linestyle = "--")

    plt.grid()
    title = ('step='+str(step)) if step!= -1 else 'True Value'
    plt.title(title)
    plt.tight_layout()
```
