```python
import matplotlib.pyplot as plt
import math

def difference_quotient(x, y):
    n = len(x)
    quotient = []
    for i in range(n):
        quotient.append(y[i])
    rank = 1
    while(rank < n):
        index_list =list(range(rank, n))
        index_list.reverse()
        for i in index_list:
            quotient[i] = (quotient[i]-quotient[i-1])/(x[i]-x[i-rank])
        rank = rank+1
    return quotient

def newton_interpolation(coefficients,x,x_test):
    result = 0
    for index,coefficient in enumerate(coefficients):
        temp = coefficient
        for i in range(index):
            temp = temp * (x_test-x[i])
        result = result+temp
    return result

def fx(x):
    return np.sin(np.pi*x)

    
plt.figure(figsize=(5, 5),dpi=200)
plt.rcParams['axes.unicode_minus']=False

for index,interval_num in enumerate([2,3,4,5,10,20]):
    ax = plt.subplot(3,2,index+1)
    x = np.arange(0,1.01,0.01)
    x_in = np.linspace(0,1,interval_num,endpoint=True)
    y_in = fx(x_in)
    quotient = difference_quotient(x_in, y_in)
    y = newton_interpolation(quotient,x_in,x)
    plt.scatter(x_in,y_in,color='blue')
    y_real = fx(x)
    plt.plot(x,y_real,color="grey",linestyle="--")
    plt.plot(x,y, color="r",linestyle = "--")

    plt.grid()
    title = ('num = '+str(interval_num)) if interval_num!= -1 else 'True Value'
    plt.title(title)
    plt.tight_layout()
```

```python
import numpy as np
import sympy

def h(x):
    h = []
    for i in range(len(x) - 1):
        h.append(x[i + 1] - x[i])
    return h

def lamda(h):
    lamda = [0]
    for i in range(1, len(h)):
        lamda.append(h[i] / (h[i] + h[i - 1]))
    return lamda

def mu(lamda):
    mu = [0]
    for i in range(1, len(lamda)):
        mu.append(1 - lamda[i])
    return mu

def g(y, mu, lamda, h, ddf_0, ddf_n):
    g = []
    g.append(3 * (y[1] - y[0]) / h[0] - h[0] / 2 * ddf_0)
    for k in range(1, len(y) - 1):
        g.append(3 * mu[k] * (y[k + 1] - y[k]) / h[k] - lamda[k] * (y[k] - y[k - 1]) / h[k - 1])
    n = len(y) - 1
    g.append(3 * (y[n] - y[n - 1]) / h[n - 1] - h[n - 1] / 2 * ddf_n)
    return g

def spline_solve_second_condition(x, y, ddf_0, ddf_n):
    n = len(x)
    h_k = h(x)
    print('h:',h_k)
    lamda_k = lamda(h_k)
    print('lambda:',lamda_k)
    mu_k = mu(lamda_k)
    print('mu:',mu_k)
    g_k = g(y, mu_k, lamda_k, h_k, ddf_0, ddf_n)
    print('g:',g_k)

    A = np.zeros((n, n))
    for i in range(1, n - 1):
        for j in range(n):
            if (i == j):
                A[i][j] = 2
            if (i == j + 1):
                A[i][j] = lamda_k[i]
            if (i == j - 1):
                A[i][j] = mu_k[i]
    A[0][1] = 1
    A[n - 1][n - 2] = 1
    A[0][0] = 2
    A[n - 1][n - 1] = 2
    print('A:',A)
    m = np.linalg.solve(A, g_k)
    print('m:',m)

    plt.figure(figsize=(5, 5),dpi=200)
    plt.rcParams['axes.unicode_minus']=False
    plt.scatter(x,y,color='blue')
    plt.grid()
    for k in range(0, n - 1):
        x_test = sympy.Symbol('x')
        s = (1 - 2 * (x_test- x[k]) / -h_k[k]) * ((x_test- x[k + 1]) / -h_k[k]) ** 2 * y[k] + (1 - 2 * (x_test- x[k + 1]) / h_k[k]) * (
                    (x_test- x[k]) / h_k[k]) ** 2 * y[k + 1] + (x_test- x[k])*((x_test- x[k + 1]) / -h_k[k]) ** 2 * m[k] + (
                        x_test- x[k + 1])*((x_test- x[k]) / h_k[k]) ** 2 * m[k + 1]
        s = sympy.expand(s)
        print("x属于[" + str(x[k]) + ", " + str(x[k + 1]) + "]时，s(x) = ", s)
        range_x = np.arange(x[k],x[k+1]+0.01,0.01)
        range_y = [s.evalf(subs={x_test:x_i}) for x_i in range_x]
        plt.plot(range_x,range_y, color="r",linestyle = "--")

x = [0, 1, 2, 3, 4]
y = [1, 3, 3, 4, 2]
ddf_0 = 0
ddf_n = 0
spline_solve_second_condition(x, y, ddf_0, ddf_n)
```
